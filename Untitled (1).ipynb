```python
def create_model(hidden_layers, units):
    model = Sequential()
    for units_count in hidden_layers:
        model.add(Dense(units=units_count, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

epochs_list = [10, 20, 30]

activation_functions = ['sigmoid', 'relu']

learning_rates = [0.1, 0.01, 0.001]

results = []
```

```python
for hidden_layers, units in [([128], []), ([128, 64], []), ([128, 64, 32], [])]:
    for epochs in epochs_list:
        for activation in activation_functions:
            for lr in learning_rates:
                
                model = create_model(hidden_layers, units)
                model.optimizer.lr = lr

                history = model.fit(x_train, y_train, epochs=epochs, validation_data=(x_val, y_val))

                results.append({
                    'hidden_layers': hidden_layers,
                    'epochs': epochs,
                    'activation': activation,
                    'learning_rate': lr,
                    'history': history.history
                })
```
